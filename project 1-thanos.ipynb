```python
import os
import random
import time
import shutil
image_lis=os.listdir("universe")
print(image_lis);
len2=len(image_lis)//2
if len(image_lis)==1:
    s=image_lis[0]
    os.remove(f"universe/{s}")
    image_lis.remove(s)
    print(f"done kill {s}")
if len(image_lis)==0:
    time.sleep(2)
    lis=os.listdir("backup")
    for i in range(len(lis)):
        shutil.copy(f"backup/{lis[i]}","universe")       
for i in range(len2):
    s=random.choice(image_lis)
    image_lis.remove(s)
    print(f"done kill {s}")
    os.remove(f"universe/{s}")
    time.sleep(2)
```
